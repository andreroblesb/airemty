```python
# load data
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
```

```python
sheets = ["SE", "CE", "SO", "NE2", "SE2", "SE3", "NE", "NO", "NO2", "NTE", "NTE2", "SO2", "SUR"]
dfs = pd.read_excel("datos_hist_2020_2025.xlsx", sheet_name=sheets)
```

```python
# Dataset dimension, number of values, number of Nans, histogram per radar
total_zones = len(dfs)
total_variables = len(dfs["SE"].columns)

nans_per_zone = {}
values_per_zone = {}
total_nans = 0
total_values = 0

for key, df in dfs.items():
    nans = df.isna().sum().sum()
    values = df.notna().sum().sum()

    nans_per_zone[key] = nans
    values_per_zone[key] = values

    total_nans += nans
    total_values += values

fig, axes = plt.subplots(2, 1, figsize=(8, 6))

# plot nans per zone
axes[0].bar(nans_per_zone.keys(), nans_per_zone.values())
axes[0].set_title("NaNs por zona")

# plot values per zone
axes[1].bar(values_per_zone.keys(), values_per_zone.values())
axes[1].set_title("Valores por zona")

plt.tight_layout()
plt.show()

print(total_nans, total_values, total_nans/total_values)
```
